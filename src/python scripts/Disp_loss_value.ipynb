```python
import numpy as np
import matplotlib.pyplot as plt

loss_values = np.load('loss_values.npy')

epochs = np.arange(1, len(loss_values) + 1)
scale = 0.7
plt.figure(figsize=(9*scale, 3.5*scale))  
plt.plot(epochs, loss_values, marker='o', linestyle='-', color='red', label='Функция потерь')  

# plt.title('Функ', fontsize=16)
plt.xlabel('Эпохи', fontsize=14)
plt.ylabel('Функция потерь', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.6)

plt.legend(fontsize=12)

# Highlight the lowest loss
min_loss = min(loss_values)
min_epoch = np.argmin(loss_values) + 1
plt.scatter(min_epoch, min_loss, color='blue', label=f'Min Loss: {min_loss:.4f}')
plt.annotate(f'Эпоха {min_epoch}\nФункция потерь {min_loss:.4f}', 
             xy=(min_epoch, min_loss), 
             xytext=(min_epoch - 5500, min_loss + 0.3), 
             arrowprops=dict(facecolor='blue', arrowstyle='->'), fontsize=14)

plt.show()
```
