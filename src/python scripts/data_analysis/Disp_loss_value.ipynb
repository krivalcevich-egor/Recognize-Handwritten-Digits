```python
import numpy as np
import matplotlib.pyplot as plt

loss_values = np.load('loss_values.npy')

epochs = np.arange(1, len(loss_values) + 1)
scale = 0.8
fig = plt.figure(figsize=(8*scale, 2.0*scale))  
plt.plot(epochs, loss_values, linestyle='-', color='red', label='Функция потерь')  # marker='o'

plt.xlabel('Номер эпохи', fontsize=11)
plt.ylabel('Функция потерь', fontsize=11)

plt.grid(True, linestyle='--', alpha=0.6)

# plt.legend(fontsize=12)

# Highlight the lowest loss
min_loss = min(loss_values)
min_epoch = np.argmin(loss_values) + 1
plt.scatter(min_epoch, min_loss, color='blue', label=f'Min Loss: {min_loss:.4f}')
plt.annotate(f'Эпоха {min_epoch}\nФункция потерь {min_loss:.4f}', 
             xy=(min_epoch, min_loss), 
             xytext=(min_epoch - 5500, min_loss + 0.3), 
             arrowprops=dict(facecolor='blue', arrowstyle='->'), fontsize=11)

plt.show()

image_name = '../../../Publication/Doklady_bsuir_2024/Figures/Loss_fn.svg'
fig.savefig(image_name, format='svg', dpi=200, bbox_inches='tight', pad_inches = 0.1)
```
